. Check the correlation between AMT_CREDIT and TARGET

. What type of loan is mostly paid late



```python
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
corrmat=pd_applications.corr()
f,ax=plt.subplots(figsize=(12,9))
sns.heatmap(corrmat,vmax=.8,square=True)
```

```python
number_of_variables=15
selected_cols=corrmat.nlargest(number_of_variables,'TARGET')['TARGET'].index
correlation_coefficient=np.corrcoef(pd_applications[selected_cols].values.T)
sns.heatmap(correlation_coefficient,vmax=.8,square=True,yticklabels=selected_cols.values,xticklabels=selected_cols.values,annot=True)
```

```python
AT_corr=pd_applications['AMT_CREDIT'].corr(pd_applications['TARGET'])
AT_corr
```

```python

target_cash_1=len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Cash loans'][pd_applications['TARGET']==1])
target_cash_2=len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Cash loans'][pd_applications['TARGET']==0])
target_rev_1=len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Resolving loans'][pd_applications['TARGET']==1])
target_rev_2=len(pd_applications[pd_applications['NAME_CONTRACT_TYPE']=='Resolving loans'][pd_applications['TARGET']==0])
x=[1,1.5]
y1,y2=[target_cash_1,target_cash_2],[target_rev_1,target_rev_2]

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

w=0.2
ax.bar(x,y1,width=w,label='Cash')
ax.bar(np.array(x)+w,y2,width=w,label='Resolving')
ax.set_xticks([1.1,1.6])
ax.set_xticklabels(['target0','target1'])
ax.legend()
plt.show()

```
